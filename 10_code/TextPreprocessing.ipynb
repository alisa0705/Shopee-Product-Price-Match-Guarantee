```python
import pandas as pd
```

```python
data = pd.read_csv('../00_source_data/train.csv')
data.head()
```

```python
import string
import re

def remove_punctuation(text):
    no_punc="".join([i for i in text if i not in string.punctuation])
    return no_punc

def tokenization(text):
    tokens = re.split('W+',text)
    return tokens

#remove punctuation
data['clean_msg']= data['title'].apply(lambda x:remove_punctuation(x))

#lower text
data['lower_msg']= data['clean_msg'].apply(lambda x: x.lower())

#tokenization
data['token_msg']= data['lower_msg'].apply(lambda x: tokenization(x))
data.head()
```
